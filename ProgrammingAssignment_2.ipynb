1. Write a Python program to convert kilometers to miles?

```python
# Take the user input
kilometers = float(input("Enter the val : "))

# Kilometers to Miles Conversion Factor
fac = 0.621371

# Converting kilometrs into Miles
miles = kilometers * fac 
 
print("{0} kilometers = {1} miles.".format(kilometers, miles))
```

2. Write a Python program to convert Celsius to Fahrenheit?

```python
# Take the user input
deg_cel = float(input("Enter the temperature val in celsius : "))

# Conversion into degree Fahrenheit

deg_fah = (deg_cel *  9/5) + 32

# Output
print("{0} degree celsius = {1} degree fahrenheit".format(deg_cel, deg_fah))
```

3. Write a Python program to display calendar?

- Pyhton has a bulit in function calendar to display the date and month of the given year as follows,

```python
import calendar

# Take the user input 
year = int(input("Enter the year : "))
month =  int(input("Enter the month : "))

# display it as output
print(calendar.month(year, month))
```

- Or we can check the each month date of a entire year as follows,

```python
# Take the user input 
year = int(input("Enter the year : "))

print(calendar.calendar(year))
```

4. Write a Python program to solve quadratic equation?
- Python has a module named cmath which performs complex square root. We import cmath module to solve quardratic equation.
- We all know that a quardatic equation has two roots as follows,
- For a equation a*x^2 + b*x + c = 0 the roots will be,
- x1 = (- b + squareroot(b^2 - 4*a*c)) / 2*a and,
- x2 = (- b - squareroot(b^2 - 4*a*c)) / 2*a

```python
import cmath

# Take the user input
a = float(input("Enter the co_eff val of x**2 : "))
b = float(input("Enter the co_eff val of x**1 : "))
c = float(input("Enter the co_eff val of x**0 : "))

# Solving for the root x1
x1 = (- b + cmath.sqrt(b**2 - 4*a*c)) / 2*a
# Solving for the root x2
x2 = (- b - cmath.sqrt(b**2 - 4*a*c)) / 2*a

# Output
print("The roots are : {0} and {1} respectively.".format(x1, x2))
```

5. Write a Python program to swap two variables without temp variable?

```python
# Take the user input
x_1 = input("Enter the val of x_1 : ")
x_2 = input("Enter the val of x_2 : ")

# Swapping the variables 
x_1, x_2 = x_2, x_1

# Make the output visible
print("The variable x_1 becomes after swapping is : {}".format(x_1))
print("The variable x_2 becomes after swapping is : {}".format(x_2))
```
