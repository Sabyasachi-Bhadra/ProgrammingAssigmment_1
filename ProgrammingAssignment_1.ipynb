2. Write a Python program to do arithmetical operations addition and division?

```python
# Take the user input
num_1 = input("Enter the first choice : ")
num_2 = input("Enter the second choice : ")

# Do Addition
add = float(num_1) + float(num_2)

# Do Division
div = float(num_1) / float(num_2)

# Make the output visible
print("The addtion of {0} and {1} is {2}".format(num_1, num_2, add))
print("The division of {0} and {1} is {2}".format(num_1, num_2, div))
```

3. Write a Python program to find the area of a triangle?

```python
# Take the user input
x_1 = float(input("Enter the first side : "))
x_2 = float(input("Enter the second side : "))
x_3 = float(input("Enter the third side : "))

# Calculation of semi perimeter
semiP = (x_1 + x_2 + x_3) / 2

# Calculation of area of the triangle
area = (semiP*(semiP - x_1) * (semiP - x_2) * (semiP - x_3)) ** 0.5

# Make the output visible
print("The area of the triangle with the given sides {0}, {1} and {2} is {3} sq.unit.".format(x_1, x_2, x_3, area))
```

4. Write a Python program to swap two variables?

```python
# Take the user input
x_1 = input("Enter the val of x_1 : ")
x_2 = input("Enter the val of x_2 : ")

# Swapping the variables 
x_1, x_2 = x_2, x_1

# Make the output visible
print("The variable x_1 becomes after swapping is : {}".format(x_1))
print("The variable x_2 becomes after swapping is : {}".format(x_2))
```

5. Write a Python program to generate a random number?

- In python we use random module to produce random number. We will start by by using the simplest method to produce a random number between 0 and 1 and will extend it by producing a list of random numbers. 

```python
# Producing random number between 0 to 1 using random module.

import random
r_1 = random.random()

print("The random number between 0 & 1 is : {}".format(r_1))
```

```python
# Producing a random number in a certain range.
import random
r_2 = random.randint(1,20)

print("The random number between 1 & 20 is : {}".format(r_2))
```

- We will produce a list of random number in a certain range by using a for loop as follows.

```python
import random
# Take the user input
x_1 = float(input("Enter the start range of producing a random number : "))
x_2 = float(input("Enter the end range of producing a random number : "))

# Let's make a empty list
r_l = []

for i in range(0,10):
  r_3 = random.randint(x_1, x_2)
  r_l.append(r_3)

print("The random list between {0} to {1} is : ".format(x_1, x_2) , r_l)
```
